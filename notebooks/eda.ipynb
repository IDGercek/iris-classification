## Graphical Comparison

### Sepal Length

```python
sns.barplot(data=df,
            x="Species",
            y="SepalLengthCm")

plt.title("Sepal Lenght Between Species")
plt.ylabel("Sepal Lenght (cm)")
plt.show()
```

### Sepal Width

```python
sns.barplot(data=df,
            x="Species",
            y="SepalWidthCm")

plt.title("Sepal Width Between Species")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

### Petal Length

```python
sns.barplot(data=df,
            x="Species",
            y="PetalLengthCm")

plt.title("Petal Lenght Between Species")
plt.ylabel("Petal Lenght (cm)")
plt.show()
```

### Petal Width

```python
sns.barplot(data=df,
            x="Species",
            y="PetalWidthCm")

plt.title("Petal Width Between Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```

Findings: 
- Generally, Iris Setosa are smaller in every aspect.
- Iris Versicolor and Iris Virginica tend to have closer values but Iris Versicolor are slightly smaller.
- It is easy to distinguish Iris Setosa from others, but seperating Iris Versicolor and Iris Virginica may be tougher.

# Comparison Between Features

## Sepals

```python
fig, axes = plt.subplots(1, 2, figsize=(20,6))

sns.regplot(data=df,
            x="SepalWidthCm",
            y="SepalLengthCm",
            ax=axes[0])

sns.lineplot(data=df,
             x="SepalWidthCm",
             y="SepalLengthCm",
             ax=axes[1])

plt.suptitle("Sepal Width and Length")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20,6))

sns.regplot(data=df[df["Species"]== "Iris-setosa"],
            x="SepalWidthCm",
            y="SepalLengthCm",
            ax=axes[0])
axes[0].set_title("Iris Setosa")

sns.regplot(data=df[df["Species"]== "Iris-versicolor"],
            x="SepalWidthCm",
            y="SepalLengthCm",
            ax=axes[1])
axes[1].set_title("Iris Versicolor")

sns.regplot(data=df[df["Species"]== "Iris-virginica"],
            x="SepalWidthCm",
            y="SepalLengthCm",
            ax=axes[2])
axes[2].set_title("Iris Virginica")

plt.suptitle("Sepal Width and Length for Each Species")
plt.show()
```

Findings:
- At the first glance, it seems like there is no correlation between sepal width and lenght.
- But when each species is analyzed individually, we can see there is indeed a correlation between sepal width and lenght

## Petals

```python
fig, axes = plt.subplots(1, 2, figsize=(20,6))

sns.regplot(data=df,
            x="PetalWidthCm",
            y="PetalLengthCm",
            ax=axes[0])

sns.lineplot(data=df,
             x="PetalWidthCm",
             y="PetalLengthCm",
             ax=axes[1])

plt.suptitle("Petal Width and Length")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20,6))

sns.regplot(data=df[df["Species"]== "Iris-setosa"],
            x="PetalWidthCm",
            y="PetalLengthCm",
            ax=axes[0])
axes[0].set_title("Iris Setosa")

sns.regplot(data=df[df["Species"]== "Iris-versicolor"],
            x="PetalWidthCm",
            y="PetalLengthCm",
            ax=axes[1])
axes[1].set_title("Iris Versicolor")

sns.regplot(data=df[df["Species"]== "Iris-virginica"],
            x="PetalWidthCm",
            y="PetalLengthCm",
            ax=axes[2])
axes[2].set_title("Iris Virginica")

plt.suptitle("Petal Width and Length for Each Species")
plt.show()
```

Finding: There is a direct correlation between petal width and height.

## Widths

```python
fig, axes = plt.subplots(1, 2, figsize=(20,6))

sns.regplot(data=df,
            x="PetalWidthCm",
            y="SepalWidthCm",
            ax=axes[0])

sns.lineplot(data=df,
             x="PetalWidthCm",
             y="SepalWidthCm",
             ax=axes[1])

plt.suptitle("Petal and Sepal Width")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20,6))

sns.regplot(data=df[df["Species"]== "Iris-setosa"],
            x="PetalWidthCm",
            y="SepalWidthCm",
            ax=axes[0])
axes[0].set_title("Iris Setosa")

sns.regplot(data=df[df["Species"]== "Iris-versicolor"],
            x="PetalWidthCm",
            y="SepalWidthCm",
            ax=axes[1])
axes[1].set_title("Iris Versicolor")

sns.regplot(data=df[df["Species"]== "Iris-virginica"],
            x="PetalWidthCm",
            y="SepalWidthCm",
            ax=axes[2])
axes[2].set_title("Iris Virginica")

plt.suptitle("Petal and Sepal Widths for Each Species")
plt.show()
```

Findings:
- At the first sight, it seems like there is no correlation between sepal and petal widths.
- But when each species is analyzed individually, we can see there is indeed a correlation between sepal and petal widths.

## Lengths

```python
fig, axes = plt.subplots(1, 2, figsize=(20,6))

sns.regplot(data=df,
            x="PetalLengthCm",
            y="SepalLengthCm",
            ax=axes[0])

sns.lineplot(data=df,
             x="PetalLengthCm",
             y="SepalLengthCm",
             ax=axes[1])

plt.suptitle("Petal and Sepal Length")
plt.show()
```

# Conclusion

Key Findings:
- Average size comparison between flowers is: Iris Virginica > Iris Versicolor > Iris Setosa
- Iris Setosa are easily distinguishable from the other two species.
- In general, Iris Versicolor are slightly smaller than Iris Virginica. 

Other Findings:
- There is a direct proportionality between petal width and petal lenght.
- There is a direct proportionality between petal lenght and sepal lenght.

- Although isn't obvious at first sight, there is a direct proportionality between sepal width and petal width
